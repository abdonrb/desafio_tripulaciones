## Modelo recomendaci√≥n KNN
(primer ejemplo sin filtrado previo)

```python
import pandas as pd
# from df_to_ml import df_to_ml
from sklearn.neighbors import NearestNeighbors

# Crear y entrenar el modelo KNN
model = NearestNeighbors(n_neighbors=5, algorithm='auto')
model.fit(df_ml)

# Funci√≥n para obtener recomendaciones dado un √≠ndice de actividad
def obtener_recomendaciones(indice_actividad, df_original, df_procesado, model, n_recommendations=5):
    """
    Obtiene recomendaciones de actividades similares a una actividad dada.
    
    Par√°metros:
    -----------
    indice_actividad : int
        √çndice de la actividad para la que se desean recomendaciones.
    df_original : DataFrame
        DataFrame original con toda la informaci√≥n.
    df_procesado : DataFrame
        DataFrame procesado usado para el modelo.
    model : NearestNeighbors
        Modelo KNN entrenado.
    n_recommendations : int, opcional (default=5)
        N√∫mero de recomendaciones a devolver.
        
    Retorna:
    --------
    DataFrame con las actividades recomendadas.
    """
    # Obtener las caracter√≠sticas de la actividad seleccionada
    actividad_seleccionada = df_procesado.iloc[[indice_actividad]]
    
    # Encontrar los vecinos m√°s cercanos
    distancias, indices = model.kneighbors(actividad_seleccionada, n_neighbors=n_recommendations+1)
    
    # Excluir la propia actividad (primera en la lista)
    indices = indices.flatten()[1:]
    distancias = distancias.flatten()[1:]
    
    # Crear DataFrame con las recomendaciones
    recomendaciones = df_original.iloc[indices].copy()
    recomendaciones['distancia'] = distancias
    
    # Ordenar por distancia (menor a mayor)
    recomendaciones = recomendaciones.sort_values('distancia')
    
    return recomendaciones
```

```python
# Ejemplo de uso
indice_ejemplo = 24  # Por ejemplo, la actividad "Chatines: club de lectura infantil"
print(df_original.loc[indice_ejemplo,['t√≠tulo', 'categor√≠a', 'd√≠a_d√≠as', 'distrito','subcategor√≠a']])

recomendaciones = obtener_recomendaciones(indice_ejemplo, df_original, df_ml, model)
print(recomendaciones[['t√≠tulo', 'categor√≠a', 'subcategor√≠a', 'd√≠a_d√≠as', 'distrito', 'distancia']])
```

He introducido una normalizaci√≥n, para aceptar may√∫sculas/ min√∫sculas, con y sin acentos.

```python
# Funciones para normalizar 
import unicodedata
def normalizar(texto):
    if not isinstance(texto, str):
        return texto
    texto = texto.lower().strip()
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')
    return texto
```

```python
# Funci√≥n de filtrado de los datos seg√∫n condiciones usuario
import pandas as pd

def filtrado_segun_condiciones(df, condiciones):
    df_filtrado = df.copy()

    # --- Funci√≥n de normalizaci√≥n ---
    def normalizar(texto):
        return str(texto).strip().lower()

    # --- Validaci√≥n obligatoria: ciudad ---
    if 'ciudad' not in condiciones or normalizar(condiciones['ciudad']) != 'madrid':
        print("üìç Ciudad no proporcionada o inv√°lida. Te mostramos las actividades en 'Madrid', ciudad en la que estamos trabajando actualmente.")
        condiciones['ciudad'] = 'Madrid'
    df_filtrado = df_filtrado[df_filtrado['ciudad'].str.lower() == 'madrid']
    
    # --- Normalizaci√≥n de tipos de columnas ---
    df_filtrado['edad_min'] = pd.to_numeric(df_filtrado['edad_min'], errors='coerce')
    df_filtrado['edad_max'] = pd.to_numeric(df_filtrado['edad_max'], errors='coerce')
    df_filtrado['fecha_inicio'] = pd.to_datetime(df_filtrado['fecha_inicio'], errors='coerce', dayfirst = True)
    df_filtrado['fecha_fin'] = pd.to_datetime(df_filtrado['fecha_fin'], errors='coerce', dayfirst = True)
    df_filtrado['precio_valor'] = pd.to_numeric(df_filtrado['precio_valor'], errors='coerce')

    # --- Filtro por rango de edad ---
    if 'rango_edad' in condiciones and condiciones['rango_edad']:
        edad_min, edad_max = condiciones['rango_edad']
        df_filtrado = df_filtrado[
            (df_filtrado['edad_max'] >= edad_min) & (df_filtrado['edad_min'] <= edad_max)
        ]

    # --- Filtro por fecha espec√≠fica ---
    if 'fecha' in condiciones and condiciones['fecha']:
        fecha_usuario = pd.to_datetime(condiciones['fecha'], dayfirst=True, errors='coerce')
        if pd.notna(fecha_usuario):
            df_filtrado = df_filtrado[
                (df_filtrado['fecha_inicio'] <= fecha_usuario) & 
                (df_filtrado['fecha_fin'] >= fecha_usuario)
            ]

    # --- Filtro por d√≠a ---
    if 'd√≠a' in condiciones and condiciones['d√≠a']:
        dia = normalizar(condiciones['d√≠a'])
        if dia == "fin de semana":
            df_filtrado = df_filtrado[df_filtrado['d√≠a_d√≠as'].apply(normalizar).isin(["s√°bado", "domingo"])]
        elif dia == "d√≠a de semana":
            df_filtrado = df_filtrado[~df_filtrado['d√≠a_d√≠as'].apply(normalizar).isin(["s√°bado", "domingo"])]
        else:
            df_filtrado = df_filtrado[df_filtrado['d√≠a_d√≠as'].apply(normalizar) == dia]
    
    # --- Filtro por periodicidad ---
    if 'periodicidad' in condiciones and condiciones['periodicidad']:
        periodicidad = normalizar(condiciones['periodicidad'])
        df_filtrado = df_filtrado[
            df_filtrado['periodicidad'].apply(lambda x: normalizar(str(x))) == periodicidad
        ]

    # --- Filtro por horario ---
    if 'horario' in condiciones and condiciones['horario']:
        horario = normalizar(condiciones['horario'])

        # Filtra solo las filas que tienen hora_inicio no nulo
        df_horas = df_filtrado[df_filtrado['hora_inicio'].notnull()].copy()
        
        # Extrae la hora como n√∫mero (por ejemplo 14.5 para 14:30)
        def hora_a_decimal(hora_str):
            try:
                h, m = map(int, hora_str.split(':'))
                return h + m / 60
            except:
                return None  # si falla la conversi√≥n, devuelve None

        df_horas['hora_decimal'] = df_horas['hora_inicio'].apply(hora_a_decimal)

        # Aplica el filtrado seg√∫n el tramo horario
        if horario == normalizar ("ma√±ana"):
            df_filtrado = df_horas[(df_horas['hora_decimal'] >= 7) & (df_horas['hora_decimal'] < 12)]
        elif horario == normalizar ("mediod√≠a"):
            df_filtrado = df_horas[(df_horas['hora_decimal'] >= 12) & (df_horas['hora_decimal'] < 15)]
        elif horario == normalizar ("tarde"):
            df_filtrado = df_horas[(df_horas['hora_decimal'] >= 15) & (df_horas['hora_decimal'] < 19)]
        elif horario == normalizar ("noche"):
            df_filtrado = df_horas[(df_horas['hora_decimal'] >= 19) | (df_horas['hora_decimal'] < 7)]

        # Elimina la columna auxiliar si no la necesitas
        df_filtrado = df_filtrado.drop(columns=['hora_decimal'], errors= 'ignore')


    # --- Filtro por estaci√≥n ---
    if 'estaci√≥n' in condiciones and condiciones['estaci√≥n']:
        estacion = normalizar(condiciones['estaci√≥n'])
        mes_inicio = df_filtrado['fecha_inicio'].dt.month
        if estacion == 'verano':
            df_filtrado = df_filtrado[mes_inicio.isin([6, 7, 8, 9])]
        elif estacion == 'invierno':
            df_filtrado = df_filtrado[mes_inicio.isin([12, 1, 2])]

    # --- Filtro por duraci√≥n ---
    if 'duraci√≥n' in condiciones and condiciones['duraci√≥n']:
        duracion_dias = (df_filtrado['fecha_fin'] - df_filtrado['fecha_inicio']).dt.days + 1
        duracion = normalizar(condiciones['duraci√≥n'])
        if duracion == "corta":
            df_filtrado = df_filtrado[duracion_dias <= 1]
        elif duracion == "media":
            df_filtrado = df_filtrado[(duracion_dias > 1) & (duracion_dias <= 7)]
        elif duracion == "larga":
            df_filtrado = df_filtrado[duracion_dias > 7]

    # --- Filtro por distrito ---
    if 'distrito' in condiciones and condiciones['distrito']:
        distrito = normalizar(condiciones['distrito'])
        df_filtrado = df_filtrado[df_filtrado['distrito'].apply(normalizar) == distrito]

    # --- Filtro por ciudad ---
    if 'ciudad' in condiciones and condiciones['ciudad']:
        ciudad = normalizar(condiciones['ciudad'])
        df_filtrado = df_filtrado[df_filtrado['ciudad'].apply(normalizar) == ciudad]

    # --- Filtro por precio ---
    if 'precio' in condiciones and condiciones['precio']:
        precio = normalizar(condiciones['precio'])
        if precio == "gratuito":
            df_filtrado = df_filtrado[df_filtrado['precio_valor'] == 0]
        elif precio == "bajo":
            df_filtrado = df_filtrado[(df_filtrado['precio_valor'] > 0) & (df_filtrado['precio_valor'] < 10)]
        elif precio == "medio":
            df_filtrado = df_filtrado[(df_filtrado['precio_valor'] >= 10) & (df_filtrado['precio_valor'] <= 30)]
        elif precio == "alto":
            df_filtrado = df_filtrado[df_filtrado['precio_valor'] > 30]

    # --- Filtro por categor√≠a ---
    if 'categor√≠a' in condiciones and condiciones['categor√≠a']:
        categoria = normalizar(condiciones['categor√≠a'])
        df_filtrado = df_filtrado[df_filtrado['categor√≠a'].apply(normalizar) == categoria]

    # --- Filtro por subcategor√≠a ---
    if 'subcategor√≠a' in condiciones and condiciones['subcategor√≠a']:
        subcategoria = normalizar(condiciones['subcategor√≠a'])
        df_filtrado = df_filtrado[df_filtrado['subcategor√≠a'].apply(normalizar) == subcategoria]

    print(df_filtrado.shape)
    return df_filtrado
```
