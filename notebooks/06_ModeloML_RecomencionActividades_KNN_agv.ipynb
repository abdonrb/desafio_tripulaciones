## Modelo recomendación KNN
(primer ejemplo sin filtrado previo)

```python
import pandas as pd
# from df_to_ml import df_to_ml
from sklearn.neighbors import NearestNeighbors

# Crear y entrenar el modelo KNN
model = NearestNeighbors(n_neighbors=5, algorithm='auto')
model.fit(df_ml)

# Función para obtener recomendaciones dado un índice de actividad
def obtener_recomendaciones(indice_actividad, df_original, df_procesado, model, n_recommendations=5):
    """
    Obtiene recomendaciones de actividades similares a una actividad dada.
    
    Parámetros:
    -----------
    indice_actividad : int
        Índice de la actividad para la que se desean recomendaciones.
    df_original : DataFrame
        DataFrame original con toda la información.
    df_procesado : DataFrame
        DataFrame procesado usado para el modelo.
    model : NearestNeighbors
        Modelo KNN entrenado.
    n_recommendations : int, opcional (default=5)
        Número de recomendaciones a devolver.
        
    Retorna:
    --------
    DataFrame con las actividades recomendadas.
    """
    # Obtener las características de la actividad seleccionada
    actividad_seleccionada = df_procesado.iloc[[indice_actividad]]
    
    # Encontrar los vecinos más cercanos
    distancias, indices = model.kneighbors(actividad_seleccionada, n_neighbors=n_recommendations+1)
    
    # Excluir la propia actividad (primera en la lista)
    indices = indices.flatten()[1:]
    distancias = distancias.flatten()[1:]
    
    # Crear DataFrame con las recomendaciones
    recomendaciones = df_original.iloc[indices].copy()
    recomendaciones['distancia'] = distancias
    
    # Ordenar por distancia (menor a mayor)
    recomendaciones = recomendaciones.sort_values('distancia')
    
    return recomendaciones
```

```python
# Ejemplo de uso
indice_ejemplo = 24  # Por ejemplo, la actividad "Chatines: club de lectura infantil"
print(df_original.loc[indice_ejemplo,['título', 'categoría', 'día_días', 'distrito','subcategoría']])

recomendaciones = obtener_recomendaciones(indice_ejemplo, df_original, df_ml, model)
print(recomendaciones[['título', 'categoría', 'subcategoría', 'día_días', 'distrito', 'distancia']])
```

Ahora introduzco una función de validación de las condiciones introducidas. Si alguna no se ha introducido, o no es valida se deja vacía. Madrid es condición obligatoria, saltaría un mensaje de error.

He introducido una normalización, para aceptar mayúsculas/ minúsculas, con y sin acentos.

```python
# Funciones para normalizar 
import unicodedata
def normalizar(texto):
    if not isinstance(texto, str):
        return texto
    texto = texto.lower().strip()
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')
    return texto
```

```python
# Función para validar condiciones introducidas por el usuario
from datetime import datetime
def validar_condiciones_usuario(condiciones_usuario):
    errores = []

    # Listas de valores válidos (sin normalizar, para referencia final)
    categorias = [
        "La Ciudad", "Cultura y Espectáculos", "Educación", "Aire Libre", "Deporte"
    ]
    subcategorias = {
        "La Ciudad": ["Actividades", "Carnaval", "Semana Santa", "Verano", "Centros de Atención a Las Familias", "Navidad"],
        "Cultura y Espectáculos": ["Cuentacuentos", "Títeres y marionetas", "Magia", "Club de lectura", "Cine Infantil", 
                                "Circo", "Conciertos", "Exposiciones", "Monumentos", "Museos", "Proyecciones", 
                                "Talleres", "Teatro Infantil", "Música", "Bailes"],
        "Educación": ["Ayudas y Becas", "Campamentos Urbanos", "Ludotecas", "Idiomas"],
        "Aire Libre": ["Actividades al Aire Libre", "Actividades con animales", "Parques y jardines"],
        "Deporte": ["Deportes de aventura", "Deportes de invierno", "Deportes acuáticos", "Deportes de cancha", 
                    "Rocódromos", "Deporte en la calle", "Carreras y eventos"]
    }
    rango_edades_validos = [(0, 3), (4, 6), (7, 9), (10, 12), (13, 15), (16, 18)]
    dias_validos = ["lunes", "martes", "miércoles", "jueves", "viernes", "sábado", "domingo", "día de semana", "fin de semana"]
    periodicidad_valida = ["diaria", "semanal", "quincenal", "mensual"]
    horarios_validos = ["mañana", "mediodía", "tarde", "noche"]
    estaciones_validas = ["verano", "invierno"]
    duraciones_validas = ["corta", "media", "larga"]
    distritos_madrid = ['Centro', 'Arganzuela', 'Retiro', 'Salamanca', 'Chamartín', 'Tetuán', 'Chamberí',
                        'Fuencarral-El Pardo', 'Moncloa-Aravaca', 'Latina', 'Carabanchel', 'Usera',
                        'Puente de Vallecas', 'Moratalaz', 'Ciudad Lineal', 'Hortaleza', 'Villaverde',
                        'Villa de Vallecas', 'Vicálvaro', 'San Blas-Canillejas', 'Barajas']
    precios_validos = ["gratuito", "bajo", "medio", "alto"]

    claves_validas = {
        'rango_edad', 'fecha', 'día', 'periodicidad', 'horario', 'estación', 'duración',
        'distrito', 'ciudad', 'precio', 'categoría', 'subcategoría'
    }

    if not condiciones_usuario:
        errores.append("Debe introducirse al menos una condición.")
        return errores

    # Detectar claves no reconocidas
    for clave in condiciones_usuario:
        if clave not in claves_validas:
            errores.append(f"Clave no reconocida: '{clave}'")

   # Validar ciudad
    ciudad = condiciones_usuario.get('ciudad', '')
    if not ciudad:
        print("📍 Ciudad no proporcionada. Te mostramos las actividades en 'Madrid', ciudad en la que estamos trabajando actualmente.")
    elif normalizar(ciudad) != 'madrid':
        print("⚠️ Ciudad inválida. Te mostramos las actividades en 'Madrid', ciudad en la que estamos trabajando actualmente.")
        condiciones_usuario ['ciudad'] = 'Madrid'


    # Validar rango de edad
    if 'rango_edad' in condiciones_usuario:
        if condiciones_usuario['rango_edad'] not in rango_edades_validos:
            errores.append(f"Rango de edad inválido: {condiciones_usuario['rango_edad']}")

    # Validar fecha
    if 'fecha' in condiciones_usuario:
        try:
            datetime.strptime(condiciones_usuario['fecha'], "%d/%m/%Y")
        except ValueError:
            errores.append(f"Formato de fecha inválido: {condiciones_usuario['fecha']} (debe ser dd/mm/aaaa)")

    # Validar día
    if 'día' in condiciones_usuario:
        dia = normalizar(condiciones_usuario['día'])
        if dia not in list(map(normalizar, dias_validos)):
            errores.append(f"Día inválido: {condiciones_usuario['día']}")
            
    # Periodicidad
    if 'periodicidad' in condiciones_usuario:
        periodicidad = normalizar(condiciones_usuario['periodicidad'])
        if periodicidad not in list(map(normalizar, periodicidad_valida)):
            errores.append(f"Periodicidad inválida: {condiciones_usuario['periodicidad']}")


    # Horario
    if 'horario' in condiciones_usuario:
        horario = normalizar(condiciones_usuario['horario'])
        if horario not in list(map(normalizar, horarios_validos)):
            errores.append(f"Horario inválido: {condiciones_usuario['horario']}")

    # Estación
    if 'estación' in condiciones_usuario and condiciones_usuario['estación']:
        estacion = normalizar(condiciones_usuario['estación'])
        if estacion not in list(map(normalizar, estaciones_validas)):
            errores.append(f"Estación inválida: {condiciones_usuario['estación']}")

    # Duración
    if 'duración' in condiciones_usuario:
        duracion = normalizar(condiciones_usuario['duración'])
        if duracion not in list(map(normalizar, duraciones_validas)):
            errores.append(f"Duración inválida: {condiciones_usuario['duración']}")

    # Distrito
    if 'distrito' in condiciones_usuario:
        distrito = normalizar(condiciones_usuario['distrito'])
        if distrito not in list(map(normalizar, distritos_madrid)):
            errores.append(f"Distrito inválido: {condiciones_usuario['distrito']}")

    # Precio
    if 'precio' in condiciones_usuario:
        precio = normalizar(condiciones_usuario['precio'])
        if precio not in list(map(normalizar, precios_validos)):
            errores.append(f"Precio inválido: {condiciones_usuario['precio']}")

    # Categoría y Subcategoría
    if 'categoría' in condiciones_usuario:
        cat_usuario = condiciones_usuario['categoría']
        cat_normal = normalizar(cat_usuario)
        cat_match = None
        for cat in categorias:
            if normalizar(cat) == cat_normal:
                cat_match = cat
                break
        if not cat_match:
            errores.append(f"Categoría inválida: {cat_usuario}")
        else:
            if 'subcategoría' in condiciones_usuario:
                subcat_usuario = condiciones_usuario['subcategoría']
                subcat_normal = normalizar(subcat_usuario)
                subcat_match = [s for s in subcategorias[cat_match] if normalizar(s) == subcat_normal]
                if not subcat_match:
                    errores.append(f"Subcategoría inválida: '{subcat_usuario}' no pertenece a la categoría '{cat_match}'")

    return True if not errores else errores
```

```python
# Función de filtrado de los datos según condiciones usuario
import pandas as pd

def filtrado_segun_condiciones(df, condiciones):
    df_filtrado = df.copy()

    # --- Función de normalización ---
    def normalizar(texto):
        return str(texto).strip().lower()

    # --- Validación obligatoria: ciudad ---
    if 'ciudad' not in condiciones or normalizar(condiciones['ciudad']) != 'madrid':
        print("📍 Ciudad no proporcionada o inválida. Te mostramos las actividades en Madrid, ciudad en la que estamos trabajando actualmente.")
        condiciones['ciudad'] = 'Madrid'
    df_filtrado = df_filtrado[df_filtrado['ciudad'].str.lower() == 'madrid']
    
    # --- Normalización de tipos de columnas ---
    df_filtrado['edad_min'] = pd.to_numeric(df_filtrado['edad_min'], errors='coerce')
    df_filtrado['edad_max'] = pd.to_numeric(df_filtrado['edad_max'], errors='coerce')
    fecha_inicio_dt = pd.to_datetime(df_filtrado['fecha_inicio'], errors='coerce', dayfirst = True)
    fecha_fin_dt = pd.to_datetime(df_filtrado['fecha_fin'], errors='coerce', dayfirst = True)
    df_filtrado['precio_valor'] = pd.to_numeric(df_filtrado['precio_valor'], errors='coerce')

    # --- Filtro por rango de edad ---
    if 'rango_edad' in condiciones and condiciones['rango_edad']:
        edad_min, edad_max = condiciones['rango_edad']
        df_filtrado = df_filtrado[
            (df_filtrado['edad_max'] >= edad_min) & (df_filtrado['edad_min'] <= edad_max)
        ]

    # --- Filtro por fecha específica ---
    if 'fecha' in condiciones and condiciones['fecha']:
        fecha_usuario = pd.to_datetime(condiciones['fecha'], dayfirst=True, errors='coerce')
        if pd.notna(fecha_usuario):
            df_filtrado = df_filtrado[
                (fecha_inicio_dt <= fecha_usuario) & 
                (fecha_fin_dt >= fecha_usuario)
            ]

    # --- Filtro por día ---
    if 'día' in condiciones and condiciones['día']:
        dia = normalizar(condiciones['día'])
        if dia == "fin de semana":
            df_filtrado = df_filtrado[df_filtrado['día_días'].apply(normalizar).isin(["sábado", "domingo"])]
        elif dia == "día de semana":
            df_filtrado = df_filtrado[~df_filtrado['día_días'].apply(normalizar).isin(["sábado", "domingo"])]
        else:
            df_filtrado = df_filtrado[df_filtrado['día_días'].apply(normalizar) == dia]
    
    # --- Filtro por periodicidad ---
    if 'periodicidad' in condiciones and condiciones['periodicidad']:
        periodicidad = normalizar(condiciones['periodicidad'])
        df_filtrado = df_filtrado[
            df_filtrado['periodicidad'].apply(lambda x: normalizar(str(x))) == periodicidad
        ]
        
    # --- Filtro por horario ---
    if 'horario' in condiciones and condiciones['horario']:
        horario = normalizar(condiciones['horario'])

        # Elimina nulos antes de convertir
        df_filtrado = df_filtrado[df_filtrado['hora_inicio'].notnull()].copy()

        def hora_a_decimal(hora_str):
            try:
                h, m = map(int, hora_str.split(':'))
                return h + m / 60
            except:
                return None

        df_filtrado['hora_decimal'] = df_filtrado['hora_inicio'].apply(hora_a_decimal)

        if horario == "mañana":
            df_filtrado = df_filtrado[(df_filtrado['hora_decimal'] >= 7) & (df_filtrado['hora_decimal'] < 12)]
        elif horario == "mediodía":
            df_filtrado = df_filtrado[(df_filtrado['hora_decimal'] >= 12) & (df_filtrado['hora_decimal'] < 15)]
        elif horario == "tarde":
            df_filtrado = df_filtrado[(df_filtrado['hora_decimal'] >= 15) & (df_filtrado['hora_decimal'] < 19)]
        elif horario == "noche":
            df_filtrado = df_filtrado[(df_filtrado['hora_decimal'] >= 19) | (df_filtrado['hora_decimal'] < 7)]

        df_filtrado = df_filtrado.drop(columns=['hora_decimal'], errors='ignore')


    # --- Filtro por estación ---
    if 'estación' in condiciones and condiciones['estación']:
        estacion = normalizar(condiciones['estación'])
        mes_inicio = fecha_inicio_dt.dt.month
        if estacion == 'verano':
            df_filtrado = df_filtrado[mes_inicio.isin([6, 7, 8, 9])]
        elif estacion == 'invierno':
            df_filtrado = df_filtrado[mes_inicio.isin([12, 1, 2])]

    # --- Filtro por duración ---
    if 'duración' in condiciones and condiciones['duración']:
        duracion_dias = (fecha_fin_dt - fecha_inicio_dt).dt.days + 1
        duracion = normalizar(condiciones['duración'])
        if duracion == "corta":
            df_filtrado = df_filtrado[duracion_dias <= 1]
        elif duracion == "media":
            df_filtrado = df_filtrado[(duracion_dias > 1) & (duracion_dias <= 7)]
        elif duracion == "larga":
            df_filtrado = df_filtrado[duracion_dias > 7]

    # --- Filtro por distrito ---
    if 'distrito' in condiciones and condiciones['distrito']:
        distrito = normalizar(condiciones['distrito'])
        df_filtrado = df_filtrado[df_filtrado['distrito'].apply(normalizar) == distrito]

    # --- Filtro por precio ---
    if 'precio' in condiciones and condiciones['precio']:
        precio = normalizar(condiciones['precio'])
        if precio == "gratuito":
            df_filtrado = df_filtrado[df_filtrado['precio_valor'] == 0]
        elif precio == "bajo":
            df_filtrado = df_filtrado[(df_filtrado['precio_valor'] > 0) & (df_filtrado['precio_valor'] < 10)]
        elif precio == "medio":
            df_filtrado = df_filtrado[(df_filtrado['precio_valor'] >= 10) & (df_filtrado['precio_valor'] <= 30)]
        elif precio == "alto":
            df_filtrado = df_filtrado[df_filtrado['precio_valor'] > 30]

    # --- Filtro por categoría ---
    if 'categoría' in condiciones and condiciones['categoría']:
        categoria = normalizar(condiciones['categoría'])
        df_filtrado = df_filtrado[df_filtrado['categoría'].apply(normalizar) == categoria]

    # --- Filtro por subcategoría ---
    if 'subcategoría' in condiciones and condiciones['subcategoría']:
        subcategoria = normalizar(condiciones['subcategoría'])
        df_filtrado = df_filtrado[df_filtrado['subcategoría'].apply(normalizar) == subcategoria]

    print(df_filtrado.shape)
    return df_filtrado
```

### Vectorización condiciones del usuario

```python
# v.1
import pandas as pd
import numpy as np
from datetime import datetime

def vectorizar_condiciones(condiciones_usuario, df_ml_columns):
    """
    Convierte las condiciones del usuario en un vector con la misma estructura que df_ml.

    Parámetros:
    -----------
    condiciones_usuario : dict
        Diccionario con las condiciones del usuario.
    df_ml_columns : list
        Lista de columnas del DataFrame df_ml para asegurar consistencia.

    Retorna:
    --------
    DataFrame con una única fila representando las condiciones del usuario.
    """
    # Inicializar un diccionario con todas las columnas de df_ml en 0
    vector_usuario = {col: 0 for col in df_ml_columns}

    # Procesar rango de edad
    if 'rango_edad' in condiciones_usuario and condiciones_usuario['rango_edad']:
        edad_min, edad_max = condiciones_usuario['rango_edad']
        rangos_edad = [(0, 3), (4, 6), (7, 9), (10, 12), (13, 15), (16, 18)]
        for min_age, max_age in rangos_edad:
            col_name = f"edad_{min_age}_{max_age}"
            if col_name in vector_usuario:
                if not (edad_max < min_age or edad_min > max_age):  # Si hay solapamiento
                    vector_usuario[col_name] = 1

    # Procesar fecha
    if 'fecha' in condiciones_usuario and condiciones_usuario['fecha']:
        try:
            fecha = datetime.strptime(condiciones_usuario['fecha'], "%d/%m/%Y")
            vector_usuario['fecha_inicio_year'] = fecha.year
            vector_usuario['fecha_inicio_month'] = fecha.month
            vector_usuario['fecha_inicio_day'] = fecha.day
            vector_usuario['fecha_inicio_dayofweek'] = fecha.weekday()
            vector_usuario['fecha_inicio_is_summer'] = int(6 <= fecha.month <= 9)
            vector_usuario['fecha_inicio_is_winter'] = int(fecha.month in [12, 1, 2])
            vector_usuario['fecha_inicio_is_weekend'] = int(fecha.weekday() >= 5)
        except ValueError:
            pass  # Ignorar si la fecha no es válida

    # Procesar día
    dias_semana = {
        'lunes': 'dia_lunes', 'martes': 'dia_martes', 'miércoles': 'dia_miércoles',
        'jueves': 'dia_jueves', 'viernes': 'dia_viernes', 'sábado': 'dia_sábado',
        'domingo': 'dia_domingo'
    }
    if 'día' in condiciones_usuario and condiciones_usuario['día']:
        dia = condiciones_usuario['día'].lower()
        if dia in dias_semana:
            vector_usuario[dias_semana[dia]] = 1
        elif dia == 'día de semana':
            for d in ['dia_lunes', 'dia_martes', 'dia_miércoles', 'dia_jueves', 'dia_viernes']:
                vector_usuario[d] = 1
        elif dia == 'fin de semana':
            vector_usuario['dia_sábado'] = 1
            vector_usuario['dia_domingo'] = 1

    # Procesar periodicidad
    if 'periodicidad' in condiciones_usuario and condiciones_usuario['periodicidad']:
        periodicidad = condiciones_usuario['periodicidad'].lower()
        vector_usuario[f'periodicidad_{periodicidad}'] = 1

    # Procesar horario
    horarios = {
        'mañana': (7, 12), 'mediodía': (12, 15), 'tarde': (15, 19), 'noche': (19, 7)
    }
    if 'horario' in condiciones_usuario and condiciones_usuario['horario']:
        horario = condiciones_usuario['horario'].lower()
        if horario in horarios:
            inicio, fin = horarios[horario]
            vector_usuario['hora_inicio'] = inicio
            vector_usuario['hora_fin'] = fin
            vector_usuario[f'hora_inicio_{horario}'] = 1

    # Procesar duración
    if 'duración' in condiciones_usuario and condiciones_usuario['duración']:
        duracion = condiciones_usuario['duración'].lower()
        vector_usuario[f'duracion_{duracion}'] = 1

    # Procesar distrito
    if 'distrito' in condiciones_usuario and condiciones_usuario['distrito']:
        distrito = f"distrito_{condiciones_usuario['distrito']}"
        if distrito in vector_usuario:
            vector_usuario[distrito] = 1
    else:
        # Si no se especifica distrito, activar todos
        for col in vector_usuario:
            if col.startswith("distrito_"):
                vector_usuario[col] = 1

    # Procesar precio
    if 'precio' in condiciones_usuario and condiciones_usuario['precio']:
        precio = f"precio_{condiciones_usuario['precio']}"
        if precio in vector_usuario:
            vector_usuario[precio] = 1

    # Procesar categoría y subcategoría
    if 'categoría' in condiciones_usuario and condiciones_usuario['categoría']:
        categoria = f"categoría_{condiciones_usuario['categoría']}"
        if categoria in vector_usuario:
            vector_usuario[categoria] = 1

    if 'subcategoría' in condiciones_usuario and condiciones_usuario['subcategoría']:
        subcategoria = f"subcategoría_{condiciones_usuario['subcategoría']}"
        if subcategoria in vector_usuario:
            vector_usuario[subcategoria] = 1

    # Ciudad (Siempre Madrid)
    vector_usuario['ciudad_Madrid'] = 1

    # Convertir en DataFrame con la misma estructura que df_ml
    return pd.DataFrame([vector_usuario], columns=df_ml_columns)
```

```python
# v.2
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import unicodedata

def normalizar_texto(texto):
    """Convierte a minúsculas y elimina acentos"""
    if isinstance(texto, str):
        texto = texto.lower()
        texto = unicodedata.normalize('NFKD', texto).encode('ascii', 'ignore').decode('utf-8')
    return texto

def vectorizar_condiciones(condiciones, df_plantilla):
    """
    Transforma las condiciones del usuario en un vector de entrada compatible con el modelo ML.
    
    Args:
        condiciones (dict): Diccionario con condiciones del usuario.
        df_plantilla (pd.DataFrame): DataFrame de referencia con todas las columnas necesarias.
        
    Returns:
        pd.DataFrame: DataFrame con una sola fila vectorizada.
    """
    # Copiar plantilla
    df_ml = df_plantilla.copy()
    df_ml[:] = 0  # Inicializar todo en cero

    # Normalizar claves y valores del diccionario
    condiciones_norm = {}
    for k, v in condiciones.items():
        k_norm = normalizar_texto(k)
        if isinstance(v, str):
            v_norm = normalizar_texto(v)
        elif isinstance(v, list):
            v_norm = [normalizar_texto(x) for x in v]
        else:
            v_norm = v
        condiciones_norm[k_norm] = v_norm

    # ====================
    # Edad
    # ====================
    edad_min = condiciones_norm.get('edad_min', 0)
    edad_max = condiciones_norm.get('edad_max', 18)
    df_ml['edad_min'] = edad_min
    df_ml['edad_max'] = edad_max

    # Activar todas las franjas de edad si no se especifica
    if 'edad_min' not in condiciones_norm and 'edad_max' not in condiciones_norm:
        for col in df_ml.columns:
            if col.startswith('edad_'):
                df_ml[col] = 1

    # ====================
    # Horario
    # ====================
    hora_inicio = condiciones_norm.get('hora_inicio', 7)
    hora_fin = condiciones_norm.get('hora_fin', 7)
    df_ml['hora_inicio'] = hora_inicio
    df_ml['hora_fin'] = hora_fin

    if 'hora_inicio' not in condiciones_norm and 'hora_fin' not in condiciones_norm:
        for col in df_ml.columns:
            if col.startswith('horario_'):
                df_ml[col] = 1

    # ====================
    # Fecha
    # ====================
    df_ml['fecha_inicio'] = condiciones_norm.get('fecha_inicio', np.nan)
    df_ml['fecha_fin'] = condiciones_norm.get('fecha_fin', np.nan)

    # ====================
    # Precio
    # ====================
    df_ml['precio_valor'] = condiciones_norm.get('precio_valor', 0)
    if 'precio_valor' not in condiciones_norm:
        for col in df_ml.columns:
            if col.startswith('precio_'):
                df_ml[col] = 1

    # ====================
    # Duración
    # ====================
    for col in df_ml.columns:
        if col.startswith('duracion_'):
            df_ml[col] = 1

    # ====================
    # Día, estación, distrito, etc.
    # ====================
    for clave, valor in condiciones_norm.items():
        if isinstance(valor, list):
            for v in valor:
                col = f"{clave}_{v}"
                if col in df_ml.columns:
                    df_ml[col] = 1
        else:
            col = f"{clave}_{valor}"
            if col in df_ml.columns:
                df_ml[col] = 1

    # ====================
    # Estandarización
    # ====================
    cols_a_escalar = ['edad_min', 'edad_max', 'hora_inicio', 'hora_fin', 'precio_valor', 'fecha_inicio', 'fecha_fin']
    for col in cols_a_escalar:
        if col in df_ml.columns and df_ml[col].notnull().any():
            df_ml[col] = StandardScaler().fit_transform(df_ml[[col]])

    # ====================
    # Rellenar NaN con 0
    # ====================
    df_ml.fillna(0, inplace=True)

    return df_ml
```

```python
# v.4
import unicodedata
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

def normalizar_texto(texto):
    if not isinstance(texto, str):
        return texto
    texto = texto.lower()
    texto = ''.join(c for c in unicodedata.normalize('NFD', texto)
                    if unicodedata.category(c) != 'Mn')
    return texto

def vectorizar_condiciones(condiciones_usuario, df_ml_columns, categorias_validas, subcategorias_validas):
    vector_usuario = pd.Series(0, index=df_ml_columns)

    condiciones_normalizadas = {}
    for key, value in condiciones_usuario.items():
        key_normalizado = key.strip().lower()
        if isinstance(value, str):
            value_normalizado = value.strip().lower()
        else:
            value_normalizado = value
        condiciones_normalizadas[key_normalizado] = value_normalizado

    # 3.1 Edad
    if 'rango_edad' in condiciones_normalizadas:
        rango_edad = condiciones_normalizadas['rango_edad']
        if isinstance(rango_edad, str) and '-' in rango_edad:
            try:
                edad_min, edad_max = map(int, rango_edad.split('-'))
            except:
                edad_min, edad_max = 0, 18
        elif isinstance(rango_edad, (list, tuple)) and len(rango_edad) == 2:
            edad_min, edad_max = rango_edad
        else:
            edad_min, edad_max = 0, 18
        vector_usuario['edad_min'] = edad_min
        vector_usuario['edad_max'] = edad_max
    else:
        vector_usuario['edad_min'] = 0
        vector_usuario['edad_max'] = 18

    # 3.2 Hora
    if 'hora_inicio' in condiciones_normalizadas:
        try:
            h = pd.to_datetime(condiciones_normalizadas['hora_inicio'], format='%H:%M', errors='coerce')
            if pd.notna(h):
                vector_usuario['hora_inicio'] = h.hour + h.minute / 60
        except:
            vector_usuario['hora_inicio'] = 7
    else:
        vector_usuario['hora_inicio'] = 7

    if 'hora_fin' in condiciones_normalizadas:
        try:
            h = pd.to_datetime(condiciones_normalizadas['hora_fin'], format='%H:%M', errors='coerce')
            if pd.notna(h):
                vector_usuario['hora_fin'] = h.hour + h.minute / 60
        except:
            vector_usuario['hora_fin'] = 7
    else:
        vector_usuario['hora_fin'] = 7

    if 'hora_inicio' not in condiciones_normalizadas or 'hora_fin' not in condiciones_normalizadas:
        for col in df_ml_columns:
            if 'hora' in col:
                vector_usuario[col] = 1

    # 3.3 Fecha
    if 'fecha' in condiciones_normalizadas:
        try:
            f = pd.to_datetime(condiciones_normalizadas['fecha'], errors='coerce', dayfirst=True)
            if pd.notna(f):
                vector_usuario['fecha_inicio'] = (f - pd.Timestamp("1970-01-01")) // pd.Timedelta('1D')
                vector_usuario['fecha_fin'] = vector_usuario['fecha_inicio']
        except:
            vector_usuario['fecha_inicio'] = (pd.Timestamp("2025-01-01") - pd.Timestamp("1970-01-01")) // pd.Timedelta('1D')
            vector_usuario['fecha_fin'] = (pd.Timestamp("2025-12-31") - pd.Timestamp("1970-01-01")) // pd.Timedelta('1D')
    else:
        vector_usuario['fecha_inicio'] = (pd.Timestamp("2025-01-01") - pd.Timestamp("1970-01-01")) // pd.Timedelta('1D')
        vector_usuario['fecha_fin'] = (pd.Timestamp("2025-12-31") - pd.Timestamp("1970-01-01")) // pd.Timedelta('1D')

    # 3.4 Precio
    if 'precio' not in condiciones_normalizadas:
        for col in df_ml_columns:
            if 'precio' in col:
                vector_usuario[col] = 1

    # 3.5 Duración
    if 'duracion' not in condiciones_normalizadas:
        for col in df_ml_columns:
            if 'duracion' in col:
                vector_usuario[col] = 1

    # 4. Escalado
    scale_cols = ['edad_min', 'edad_max', 'hora_inicio', 'hora_fin', 'precio_valor', 'fecha_inicio', 'fecha_fin']
    scaler = StandardScaler()
    for col in scale_cols:
        if col in vector_usuario:
            valor = vector_usuario[[col]].values.reshape(-1, 1)  # Extraer como 2D array
            vector_usuario[col] = scaler.fit_transform(valor).flatten()  # Asignar como array 1D


    # 5. Rellenar NaNs
    vector_usuario.fillna(0, inplace=True)

    return vector_usuario
```

```python
# v.2 función recomendar con modo 'soft' ponderando condiciones clave
# condiciones clave son 'ciudad''distrito''rango_edad''fecha'
def recomendar_actividades(condiciones_usuario, modo='hard', k=5, 
                           df_original=None, df_ml=None, 
                           modelo_knn_filtrado=None, modelo_knn_global=None,
                           df_filtrado=None, df_filtrado_ml=None):
    
    validacion = validar_condiciones_usuario(condiciones_usuario)
    if validacion is not True:
        return {"status": "error", "errores": validacion}

    try:
        # Usamos un vector de ejemplo para el usuario (esto debería venir de tu vectorización real)
        # Asegúrate de usar vectorizar_condiciones cuando tengas la implementación real
        vector_usuario = df_ml.iloc[[25]]  # Esto debe ser reemplazado con la vectorización real
    except Exception as e:
        return {"status": "error", "mensaje": f"Error al vectorizar condiciones: {e}"}

    if modo == 'hard':
        # Asegúrate de usar los datos filtrados y el modelo filtrado
        if df_filtrado_ml is None or modelo_knn_filtrado is None or df_filtrado is None:
            return {"status": "error", "mensaje": "Faltan parámetros necesarios para modo 'hard'."}
        
        # Verificar si hay suficientes actividades para recomendar
        if len(df_filtrado_ml) < k:
            mensaje = f"Solo hay {len(df_filtrado_ml)} actividades disponibles (menos que k={k}). Se muestran todas."
            k = len(df_filtrado_ml)
        else:
            mensaje = f"Modo: {modo} - Recomendación generada correctamente con {k} vecinos más cercanos."
            
        # Convertimos el vector del usuario en un DataFrame con las columnas correctas
        vector_df = pd.DataFrame([vector_usuario.values[0]], columns=df_ml.columns)
        
        # Obtener las distancias y los índices con el modelo filtrado
        distancias, indices = modelo_knn_filtrado.kneighbors(vector_df, n_neighbors=k)
        
        # Obtener los resultados usando los índices y añadir las distancias
        resultados = df_filtrado.iloc[indices[0]].copy()
        resultados["distancia"] = distancias[0]
        
        # Imprimir un mensaje de estado
        print(mensaje)
        
        # Obtener los resultados usando los índices y añadir las distancias
        resultados = df_filtrado.iloc[indices[0]].copy()
        resultados["distancia"] = distancias[0]
        
        # Retornar directamente el DataFrame con los resultados
        return resultados

    elif modo == 'soft':
        if df_ml is None or modelo_knn_global is None or df_original is None:
            return {"status": "error", "mensaje": "Faltan parámetros necesarios para modo 'soft'."}
        
        df_candidato = df_ml.copy()
        df_candidato_original = df_original.copy()
        condiciones = condiciones_usuario.copy()
        
        # Ciudad
        if condiciones.get("ciudad") == "Madrid":
            df_candidato = df_candidato[df_original["ciudad"] == "Madrid"]
            df_candidato_original = df_candidato_original[df_original["ciudad"] == "Madrid"]

        # Distrito
        if "distrito" in condiciones:
            distrito = condiciones["distrito"]
            df_candidato = df_candidato[df_original["distrito"] == distrito]
            df_candidato_original = df_candidato_original[df_original["distrito"] == distrito]

        # Rango de edad
        if "rango_edad" in condiciones:
            edad_min, edad_max = condiciones["rango_edad"]
            df_candidato = df_candidato[
                (df_original["edad_min"] <= edad_max) & 
                (df_original["edad_max"] >= edad_min)
            ]
            df_candidato_original = df_candidato_original[
                (df_original["edad_min"] <= edad_max) & 
                (df_original["edad_max"] >= edad_min)
            ]

        # Fecha
        if "fecha" in condiciones:
            fecha = condiciones["fecha"]
            df_candidato = df_candidato[df_original["fecha_inicio"] <= fecha]
            df_candidato = df_candidato[df_original["fecha_fin"] >= fecha]
            df_candidato_original = df_candidato_original[
                (df_original["fecha_inicio"] <= fecha) &
                (df_original["fecha_fin"] >= fecha)
            ]

        # Si hay suficientes actividades candidatas, usamos ese subconjunto
        if len(df_candidato) >= k:
            vector_df = pd.DataFrame([vector_usuario.values[0]], columns=df_ml.columns)
            distancias, indices = modelo_knn_global.kneighbors(vector_df, n_neighbors=k)
            resultados = df_candidato_original.iloc[indices[0]].copy()
            resultados["distancia"] = distancias[0]
            print(f"Modo: {modo} - Recomendación generada con filtrado suave y {k} vecinos.")
            return resultados

        # Fallback global si no hay suficientes
        vector_df = pd.DataFrame([vector_usuario.values[0]], columns=df_ml.columns)
        distancias, indices = modelo_knn_global.kneighbors(vector_df, n_neighbors=k)
        resultados = df_original.iloc[indices[0]].copy()
        resultados["distancia"] = distancias[0]
        print(f"Modo: {modo} - Recomendación generada sin filtrado (fallback global) con {k} vecinos.")
        return resultados

        
    else:
        # Si el modo no es válido, devolver un mensaje de error
        print("Modo no válido. Usa 'hard' o 'soft'.")
        return None
```
