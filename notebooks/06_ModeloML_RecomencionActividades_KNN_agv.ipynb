## Modelo recomendaci칩n KNN
(primer ejemplo sin filtrado previo)

```python
import pandas as pd
# from df_to_ml import df_to_ml
from sklearn.neighbors import NearestNeighbors

# Crear y entrenar el modelo KNN
model = NearestNeighbors(n_neighbors=5, algorithm='auto')
model.fit(df_ml)

# Funci칩n para obtener recomendaciones dado un 칤ndice de actividad
def obtener_recomendaciones(indice_actividad, df_original, df_procesado, model, n_recommendations=5):
    """
    Obtiene recomendaciones de actividades similares a una actividad dada.
    
    Par치metros:
    -----------
    indice_actividad : int
        칈ndice de la actividad para la que se desean recomendaciones.
    df_original : DataFrame
        DataFrame original con toda la informaci칩n.
    df_procesado : DataFrame
        DataFrame procesado usado para el modelo.
    model : NearestNeighbors
        Modelo KNN entrenado.
    n_recommendations : int, opcional (default=5)
        N칰mero de recomendaciones a devolver.
        
    Retorna:
    --------
    DataFrame con las actividades recomendadas.
    """
    # Obtener las caracter칤sticas de la actividad seleccionada
    actividad_seleccionada = df_procesado.iloc[[indice_actividad]]
    
    # Encontrar los vecinos m치s cercanos
    distancias, indices = model.kneighbors(actividad_seleccionada, n_neighbors=n_recommendations+1)
    
    # Excluir la propia actividad (primera en la lista)
    indices = indices.flatten()[1:]
    distancias = distancias.flatten()[1:]
    
    # Crear DataFrame con las recomendaciones
    recomendaciones = df_original.iloc[indices].copy()
    recomendaciones['distancia'] = distancias
    
    # Ordenar por distancia (menor a mayor)
    recomendaciones = recomendaciones.sort_values('distancia')
    
    return recomendaciones
```

```python
# Ejemplo de uso
indice_ejemplo = 24  # Por ejemplo, la actividad "Chatines: club de lectura infantil"
print(df_original.loc[indice_ejemplo,['t칤tulo', 'categor칤a', 'd칤a_d칤as', 'distrito','subcategor칤a']])

recomendaciones = obtener_recomendaciones(indice_ejemplo, df_original, df_ml, model)
print(recomendaciones[['t칤tulo', 'categor칤a', 'subcategor칤a', 'd칤a_d칤as', 'distrito', 'distancia']])
```

Ahora introduzco una funci칩n de validaci칩n de las condiciones introducidas. Si alguna no se ha introducido, o no es valida se deja vac칤a. Madrid es condici칩n obligatoria, saltar칤a un mensaje de error.

He introducido una normalizaci칩n, para aceptar may칰sculas/ min칰sculas, con y sin acentos.

```python
# Funciones para normalizar y validar condiciones introducidas por el usuario
import re
import unicodedata
from datetime import datetime

def normalizar(texto):
    if not isinstance(texto, str):
        return texto
    texto = texto.lower().strip()
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')
    return texto

def validar_condiciones_usuario(condiciones_usuario):
    errores = []

    # Listas de valores v치lidos (sin normalizar, para referencia final)
    categorias = [
        "La Ciudad", "Cultura y Espect치culos", "Educaci칩n", "Aire Libre", "Deporte"
    ]
    subcategorias = {
        "La Ciudad": ["Actividades", "Carnaval", "Semana Santa", "Verano", "Centros de Atenci칩n a Las Familias", "Navidad"],
        "Cultura y Espect치culos": ["Cuentacuentos", "T칤teres y marionetas", "Magia", "Club de lectura", "Cine Infantil", 
                                "Circo", "Conciertos", "Exposiciones", "Monumentos", "Museos", "Proyecciones", 
                                "Talleres", "Teatro Infantil", "M칰sica", "Bailes"],
        "Educaci칩n": ["Ayudas y Becas", "Campamentos Urbanos", "Ludotecas", "Idiomas"],
        "Aire Libre": ["Actividades al Aire Libre", "Actividades con animales", "Parques y jardines"],
        "Deporte": ["Deportes de aventura", "Deportes de invierno", "Deportes acu치ticos", "Deportes de cancha", 
                    "Roc칩dromos", "Deporte en la calle", "Carreras y eventos"]
    }
    rango_edades_validos = [(0, 3), (4, 6), (7, 9), (10, 12), (13, 15), (16, 18)]
    dias_validos = ["lunes", "martes", "mi칠rcoles", "jueves", "viernes", "s치bado", "domingo", "d칤a de semana", "fin de semana"]
    horarios_validos = ["ma침ana", "mediod칤a", "tarde", "noche"]
    estaciones_validas = ["verano", "invierno"]
    duraciones_validas = ["corta", "media", "larga"]
    distritos_madrid = ['Centro', 'Arganzuela', 'Retiro', 'Salamanca', 'Chamart칤n', 'Tetu치n', 'Chamber칤',
                        'Fuencarral-El Pardo', 'Moncloa-Aravaca', 'Latina', 'Carabanchel', 'Usera',
                        'Puente de Vallecas', 'Moratalaz', 'Ciudad Lineal', 'Hortaleza', 'Villaverde',
                        'Villa de Vallecas', 'Vic치lvaro', 'San Blas-Canillejas', 'Barajas']
    precios_validos = ["gratuito", "no gratuito"]

    claves_validas = {
        'rango_edad', 'fecha', 'd칤a', 'horario', 'estaci칩n', 'duraci칩n',
        'distrito', 'ciudad', 'precio', 'categor칤a', 'subcategor칤a'
    }

    if not condiciones_usuario:
        errores.append("Debe introducirse al menos una condici칩n.")
        return errores

    # Detectar claves no reconocidas
    for clave in condiciones_usuario:
        if clave not in claves_validas:
            errores.append(f"Clave no reconocida: '{clave}'")

    # Validar ciudad
    ciudad = condiciones_usuario.get('ciudad', '')
    if normalizar(ciudad) != 'madrid':
        errores.append("丘멆잺 Ciudad inv치lida o no proporcionada. Por ahora, solo podemos proponerte actividades en 'Madrid'.")

    # Validar rango de edad
    if 'rango_edad' in condiciones_usuario:
        if condiciones_usuario['rango_edad'] not in rango_edades_validos:
            errores.append(f"Rango de edad inv치lido: {condiciones_usuario['rango_edad']}")

    # Validar fecha
    if 'fecha' in condiciones_usuario:
        try:
            datetime.strptime(condiciones_usuario['fecha'], "%d/%m/%Y")
        except ValueError:
            errores.append(f"Formato de fecha inv치lido: {condiciones_usuario['fecha']} (debe ser dd/mm/aaaa)")

    # Validar d칤a
    if 'd칤a' in condiciones_usuario:
        dia = normalizar(condiciones_usuario['d칤a'])
        if dia not in list(map(normalizar, dias_validos)):
            errores.append(f"D칤a inv치lido: {condiciones_usuario['d칤a']}")

    # Horario
    if 'horario' in condiciones_usuario:
        horario = normalizar(condiciones_usuario['horario'])
        if horario not in list(map(normalizar, horarios_validos)):
            errores.append(f"Horario inv치lido: {condiciones_usuario['horario']}")

    # Estaci칩n
    if 'estaci칩n' in condiciones_usuario and condiciones_usuario['estaci칩n']:
        estacion = normalizar(condiciones_usuario['estaci칩n'])
        if estacion not in list(map(normalizar, estaciones_validas)):
            errores.append(f"Estaci칩n inv치lida: {condiciones_usuario['estaci칩n']}")

    # Duraci칩n
    if 'duraci칩n' in condiciones_usuario:
        duracion = normalizar(condiciones_usuario['duraci칩n'])
        if duracion not in list(map(normalizar, duraciones_validas)):
            errores.append(f"Duraci칩n inv치lida: {condiciones_usuario['duraci칩n']}")

    # Distrito
    if 'distrito' in condiciones_usuario:
        distrito = normalizar(condiciones_usuario['distrito'])
        if distrito not in list(map(normalizar, distritos_madrid)):
            errores.append(f"Distrito inv치lido: {condiciones_usuario['distrito']}")

    # Precio
    if 'precio' in condiciones_usuario:
        precio = normalizar(condiciones_usuario['precio'])
        if precio not in list(map(normalizar, precios_validos)):
            errores.append(f"Precio inv치lido: {condiciones_usuario['precio']}")

    # Categor칤a y Subcategor칤a
    if 'categor칤a' in condiciones_usuario:
        cat_usuario = condiciones_usuario['categor칤a']
        cat_normal = normalizar(cat_usuario)
        cat_match = None
        for cat in categorias:
            if normalizar(cat) == cat_normal:
                cat_match = cat
                break
        if not cat_match:
            errores.append(f"Categor칤a inv치lida: {cat_usuario}")
        else:
            if 'subcategor칤a' in condiciones_usuario:
                subcat_usuario = condiciones_usuario['subcategor칤a']
                subcat_normal = normalizar(subcat_usuario)
                subcat_match = [s for s in subcategorias[cat_match] if normalizar(s) == subcat_normal]
                if not subcat_match:
                    errores.append(f"Subcategor칤a inv치lida: '{subcat_usuario}' no pertenece a la categor칤a '{cat_match}'")

    return True if not errores else errores
```
