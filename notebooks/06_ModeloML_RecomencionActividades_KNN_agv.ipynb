## Modelo recomendación KNN
(primer ejemplo sin filtrado previo)

```python
import pandas as pd
# from df_to_ml import df_to_ml
from sklearn.neighbors import NearestNeighbors

# Crear y entrenar el modelo KNN
model = NearestNeighbors(n_neighbors=5, algorithm='auto')
model.fit(df_ml)

# Función para obtener recomendaciones dado un índice de actividad
def obtener_recomendaciones(indice_actividad, df_original, df_procesado, model, n_recommendations=5):
    """
    Obtiene recomendaciones de actividades similares a una actividad dada.
    
    Parámetros:
    -----------
    indice_actividad : int
        Índice de la actividad para la que se desean recomendaciones.
    df_original : DataFrame
        DataFrame original con toda la información.
    df_procesado : DataFrame
        DataFrame procesado usado para el modelo.
    model : NearestNeighbors
        Modelo KNN entrenado.
    n_recommendations : int, opcional (default=5)
        Número de recomendaciones a devolver.
        
    Retorna:
    --------
    DataFrame con las actividades recomendadas.
    """
    # Obtener las características de la actividad seleccionada
    actividad_seleccionada = df_procesado.iloc[[indice_actividad]]
    
    # Encontrar los vecinos más cercanos
    distancias, indices = model.kneighbors(actividad_seleccionada, n_neighbors=n_recommendations+1)
    
    # Excluir la propia actividad (primera en la lista)
    indices = indices.flatten()[1:]
    distancias = distancias.flatten()[1:]
    
    # Crear DataFrame con las recomendaciones
    recomendaciones = df_original.iloc[indices].copy()
    recomendaciones['distancia'] = distancias
    
    # Ordenar por distancia (menor a mayor)
    recomendaciones = recomendaciones.sort_values('distancia')
    
    return recomendaciones
```

```python
# Ejemplo de uso
indice_ejemplo = 24  # Por ejemplo, la actividad "Chatines: club de lectura infantil"
print(df_original.loc[indice_ejemplo,['título', 'categoría', 'día_días', 'distrito','subcategoría']])

recomendaciones = obtener_recomendaciones(indice_ejemplo, df_original, df_ml, model)
print(recomendaciones[['título', 'categoría', 'subcategoría', 'día_días', 'distrito', 'distancia']])
```

He introducido una normalización, para aceptar mayúsculas/ minúsculas, con y sin acentos.

```python
# Funciones para normalizar 
import unicodedata
def normalizar(texto):
    if not isinstance(texto, str):
        return texto
    texto = texto.lower().strip()
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')
    return texto
```

```python
# Función para validar condiciones introducidas por el usuario
from datetime import datetime
def validar_condiciones_usuario(condiciones_usuario):
    errores = []

    # Listas de valores válidos (sin normalizar, para referencia final)
    categorias = [
        "La Ciudad", "Cultura y Espectáculos", "Educación", "Aire Libre", "Deporte"
    ]
    subcategorias = {
        "La Ciudad": ["Actividades", "Carnaval", "Semana Santa", "Verano", "Centros de Atención a Las Familias", "Navidad"],
        "Cultura y Espectáculos": ["Cuentacuentos", "Títeres y marionetas", "Magia", "Club de lectura", "Cine Infantil", 
                                "Circo", "Conciertos", "Exposiciones", "Monumentos", "Museos", "Proyecciones", 
                                "Talleres", "Teatro Infantil", "Música", "Bailes"],
        "Educación": ["Ayudas y Becas", "Campamentos Urbanos", "Ludotecas", "Idiomas"],
        "Aire Libre": ["Actividades al Aire Libre", "Actividades con animales", "Parques y jardines"],
        "Deporte": ["Deportes de aventura", "Deportes de invierno", "Deportes acuáticos", "Deportes de cancha", 
                    "Rocódromos", "Deporte en la calle", "Carreras y eventos"]
    }
    rango_edades_validos = [(0, 3), (4, 6), (7, 9), (10, 12), (13, 15), (16, 18)]
    dias_validos = ["lunes", "martes", "miércoles", "jueves", "viernes", "sábado", "domingo", "día de semana", "fin de semana"]
    periodicidad_valida = ["diaria", "semanal", "quincenal", "mensual"]
    horarios_validos = ["mañana", "mediodía", "tarde", "noche"]
    estaciones_validas = ["verano", "invierno"]
    duraciones_validas = ["corta", "media", "larga"]
    distritos_madrid = ['Centro', 'Arganzuela', 'Retiro', 'Salamanca', 'Chamartín', 'Tetuán', 'Chamberí',
                        'Fuencarral-El Pardo', 'Moncloa-Aravaca', 'Latina', 'Carabanchel', 'Usera',
                        'Puente de Vallecas', 'Moratalaz', 'Ciudad Lineal', 'Hortaleza', 'Villaverde',
                        'Villa de Vallecas', 'Vicálvaro', 'San Blas-Canillejas', 'Barajas']
    precios_validos = ["gratuito", "bajo", "medio", "alto"]

    claves_validas = {
        'rango_edad', 'fecha', 'día', 'periodicidad', 'horario', 'estación', 'duración',
        'distrito', 'ciudad', 'precio', 'categoría', 'subcategoría'
    }

    if not condiciones_usuario:
        errores.append("Debe introducirse al menos una condición.")
        return errores

    # Detectar claves no reconocidas
    for clave in condiciones_usuario:
        if clave not in claves_validas:
            errores.append(f"Clave no reconocida: '{clave}'")

   # Validar ciudad
    ciudad = condiciones_usuario.get('ciudad', '')
    if not ciudad:
        print("📍 Ciudad no proporcionada. Te mostramos las actividades en 'Madrid', ciudad en la que estamos trabajando actualmente.")
    elif normalizar(ciudad) != 'madrid':
        print("⚠️ Ciudad inválida. Te mostramos las actividades en 'Madrid', ciudad en la que estamos trabajando actualmente.")
        condiciones_usuario ['ciudad'] = 'Madrid'


    # Validar rango de edad
    if 'rango_edad' in condiciones_usuario:
        if condiciones_usuario['rango_edad'] not in rango_edades_validos:
            errores.append(f"Rango de edad inválido: {condiciones_usuario['rango_edad']}")

    # Validar fecha
    if 'fecha' in condiciones_usuario:
        try:
            datetime.strptime(condiciones_usuario['fecha'], "%d/%m/%Y")
        except ValueError:
            errores.append(f"Formato de fecha inválido: {condiciones_usuario['fecha']} (debe ser dd/mm/aaaa)")

    # Validar día
    if 'día' in condiciones_usuario:
        dia = normalizar(condiciones_usuario['día'])
        if dia not in list(map(normalizar, dias_validos)):
            errores.append(f"Día inválido: {condiciones_usuario['día']}")
            
    # Periodicidad
    if 'periodicidad' in condiciones_usuario:
        periodicidad = normalizar(condiciones_usuario['periodicidad'])
        if periodicidad not in list(map(normalizar, periodicidad_valida)):
            errores.append(f"Periodicidad inválida: {condiciones_usuario['periodicidad']}")


    # Horario
    if 'horario' in condiciones_usuario:
        horario = normalizar(condiciones_usuario['horario'])
        if horario not in list(map(normalizar, horarios_validos)):
            errores.append(f"Horario inválido: {condiciones_usuario['horario']}")

    # Estación
    if 'estación' in condiciones_usuario and condiciones_usuario['estación']:
        estacion = normalizar(condiciones_usuario['estación'])
        if estacion not in list(map(normalizar, estaciones_validas)):
            errores.append(f"Estación inválida: {condiciones_usuario['estación']}")

    # Duración
    if 'duración' in condiciones_usuario:
        duracion = normalizar(condiciones_usuario['duración'])
        if duracion not in list(map(normalizar, duraciones_validas)):
            errores.append(f"Duración inválida: {condiciones_usuario['duración']}")

    # Distrito
    if 'distrito' in condiciones_usuario:
        distrito = normalizar(condiciones_usuario['distrito'])
        if distrito not in list(map(normalizar, distritos_madrid)):
            errores.append(f"Distrito inválido: {condiciones_usuario['distrito']}")

    # Precio
    if 'precio' in condiciones_usuario:
        precio = normalizar(condiciones_usuario['precio'])
        if precio not in list(map(normalizar, precios_validos)):
            errores.append(f"Precio inválido: {condiciones_usuario['precio']}")

    # Categoría y Subcategoría
    if 'categoría' in condiciones_usuario:
        cat_usuario = condiciones_usuario['categoría']
        cat_normal = normalizar(cat_usuario)
        cat_match = None
        for cat in categorias:
            if normalizar(cat) == cat_normal:
                cat_match = cat
                break
        if not cat_match:
            errores.append(f"Categoría inválida: {cat_usuario}")
        else:
            if 'subcategoría' in condiciones_usuario:
                subcat_usuario = condiciones_usuario['subcategoría']
                subcat_normal = normalizar(subcat_usuario)
                subcat_match = [s for s in subcategorias[cat_match] if normalizar(s) == subcat_normal]
                if not subcat_match:
                    errores.append(f"Subcategoría inválida: '{subcat_usuario}' no pertenece a la categoría '{cat_match}'")

    return True if not errores else errores
```

```python
# Función de filtrado de los datos según condiciones usuario
import pandas as pd

def filtrado_segun_condiciones(df, condiciones):
    df_filtrado = df.copy()

    # --- Función de normalización ---
    def normalizar(texto):
        return str(texto).strip().lower()

    # --- Validación obligatoria: ciudad ---
    if 'ciudad' not in condiciones or normalizar(condiciones['ciudad']) != 'madrid':
        print("📍 Ciudad no proporcionada o inválida. Te mostramos las actividades en 'Madrid', ciudad en la que estamos trabajando actualmente.")
        condiciones['ciudad'] = 'Madrid'
    df_filtrado = df_filtrado[df_filtrado['ciudad'].str.lower() == 'madrid']
    
    # --- Normalización de tipos de columnas ---
    df_filtrado['edad_min'] = pd.to_numeric(df_filtrado['edad_min'], errors='coerce')
    df_filtrado['edad_max'] = pd.to_numeric(df_filtrado['edad_max'], errors='coerce')
    df_filtrado['fecha_inicio'] = pd.to_datetime(df_filtrado['fecha_inicio'], errors='coerce', dayfirst = True)
    df_filtrado['fecha_fin'] = pd.to_datetime(df_filtrado['fecha_fin'], errors='coerce', dayfirst = True)
    df_filtrado['precio_valor'] = pd.to_numeric(df_filtrado['precio_valor'], errors='coerce')

    # --- Filtro por rango de edad ---
    if 'rango_edad' in condiciones and condiciones['rango_edad']:
        edad_min, edad_max = condiciones['rango_edad']
        df_filtrado = df_filtrado[
            (df_filtrado['edad_max'] >= edad_min) & (df_filtrado['edad_min'] <= edad_max)
        ]

    # --- Filtro por fecha específica ---
    if 'fecha' in condiciones and condiciones['fecha']:
        fecha_usuario = pd.to_datetime(condiciones['fecha'], dayfirst=True, errors='coerce')
        if pd.notna(fecha_usuario):
            df_filtrado = df_filtrado[
                (df_filtrado['fecha_inicio'] <= fecha_usuario) & 
                (df_filtrado['fecha_fin'] >= fecha_usuario)
            ]

    # --- Filtro por día ---
    if 'día' in condiciones and condiciones['día']:
        dia = normalizar(condiciones['día'])
        if dia == "fin de semana":
            df_filtrado = df_filtrado[df_filtrado['día_días'].apply(normalizar).isin(["sábado", "domingo"])]
        elif dia == "día de semana":
            df_filtrado = df_filtrado[~df_filtrado['día_días'].apply(normalizar).isin(["sábado", "domingo"])]
        else:
            df_filtrado = df_filtrado[df_filtrado['día_días'].apply(normalizar) == dia]
    
    # --- Filtro por periodicidad ---
    if 'periodicidad' in condiciones and condiciones['periodicidad']:
        periodicidad = normalizar(condiciones['periodicidad'])
        df_filtrado = df_filtrado[
            df_filtrado['periodicidad'].apply(lambda x: normalizar(str(x))) == periodicidad
        ]

    # --- Filtro por horario ---
    if 'horario' in condiciones and condiciones['horario']:
        horario = normalizar(condiciones['horario'])

        # Filtra solo las filas que tienen hora_inicio no nulo
        df_horas = df_filtrado[df_filtrado['hora_inicio'].notnull()].copy()
        
        # Extrae la hora como número (por ejemplo 14.5 para 14:30)
        def hora_a_decimal(hora_str):
            try:
                h, m = map(int, hora_str.split(':'))
                return h + m / 60
            except:
                return None  # si falla la conversión, devuelve None

        df_horas['hora_decimal'] = df_horas['hora_inicio'].apply(hora_a_decimal)

        # Aplica el filtrado según el tramo horario
        if horario == normalizar ("mañana"):
            df_filtrado = df_horas[(df_horas['hora_decimal'] >= 7) & (df_horas['hora_decimal'] < 12)]
        elif horario == normalizar ("mediodía"):
            df_filtrado = df_horas[(df_horas['hora_decimal'] >= 12) & (df_horas['hora_decimal'] < 15)]
        elif horario == normalizar ("tarde"):
            df_filtrado = df_horas[(df_horas['hora_decimal'] >= 15) & (df_horas['hora_decimal'] < 19)]
        elif horario == normalizar ("noche"):
            df_filtrado = df_horas[(df_horas['hora_decimal'] >= 19) | (df_horas['hora_decimal'] < 7)]

        # Elimina la columna auxiliar si no la necesitas
        df_filtrado = df_filtrado.drop(columns=['hora_decimal'], errors= 'ignore')


    # --- Filtro por estación ---
    if 'estación' in condiciones and condiciones['estación']:
        estacion = normalizar(condiciones['estación'])
        mes_inicio = df_filtrado['fecha_inicio'].dt.month
        if estacion == 'verano':
            df_filtrado = df_filtrado[mes_inicio.isin([6, 7, 8, 9])]
        elif estacion == 'invierno':
            df_filtrado = df_filtrado[mes_inicio.isin([12, 1, 2])]

    # --- Filtro por duración ---
    if 'duración' in condiciones and condiciones['duración']:
        duracion_dias = (df_filtrado['fecha_fin'] - df_filtrado['fecha_inicio']).dt.days + 1
        duracion = normalizar(condiciones['duración'])
        if duracion == "corta":
            df_filtrado = df_filtrado[duracion_dias <= 1]
        elif duracion == "media":
            df_filtrado = df_filtrado[(duracion_dias > 1) & (duracion_dias <= 7)]
        elif duracion == "larga":
            df_filtrado = df_filtrado[duracion_dias > 7]

    # --- Filtro por distrito ---
    if 'distrito' in condiciones and condiciones['distrito']:
        distrito = normalizar(condiciones['distrito'])
        df_filtrado = df_filtrado[df_filtrado['distrito'].apply(normalizar) == distrito]

    # --- Filtro por ciudad ---
    if 'ciudad' in condiciones and condiciones['ciudad']:
        ciudad = normalizar(condiciones['ciudad'])
        df_filtrado = df_filtrado[df_filtrado['ciudad'].apply(normalizar) == ciudad]

    # --- Filtro por precio ---
    if 'precio' in condiciones and condiciones['precio']:
        precio = normalizar(condiciones['precio'])
        if precio == "gratuito":
            df_filtrado = df_filtrado[df_filtrado['precio_valor'] == 0]
        elif precio == "bajo":
            df_filtrado = df_filtrado[(df_filtrado['precio_valor'] > 0) & (df_filtrado['precio_valor'] < 10)]
        elif precio == "medio":
            df_filtrado = df_filtrado[(df_filtrado['precio_valor'] >= 10) & (df_filtrado['precio_valor'] <= 30)]
        elif precio == "alto":
            df_filtrado = df_filtrado[df_filtrado['precio_valor'] > 30]

    # --- Filtro por categoría ---
    if 'categoría' in condiciones and condiciones['categoría']:
        categoria = normalizar(condiciones['categoría'])
        df_filtrado = df_filtrado[df_filtrado['categoría'].apply(normalizar) == categoria]

    # --- Filtro por subcategoría ---
    if 'subcategoría' in condiciones and condiciones['subcategoría']:
        subcategoria = normalizar(condiciones['subcategoría'])
        df_filtrado = df_filtrado[df_filtrado['subcategoría'].apply(normalizar) == subcategoria]

    print(df_filtrado.shape)
    return df_filtrado
```

## Recomendación con modelo KNN diseñado

Para mejorar, en la función de uso, incorporo una variable que permita recomendar en diferente modo:
- hard o estrictas, es decir, solo devolver recomendación de actividades que las cumplan todas rigurosamente (o ninguna si no hay) ó 
- soft o flexibles, es decir, que pueda devolver actividades "próximas", que no cumplan estrictamente todas las condiciones

```python
def recomendar_actividades(condiciones_usuario, modo='hard', k=5):
    validacion = validar_condiciones_usuario(condiciones_usuario)
    if validacion is not True:
        return validacion  # lista de errores

    if modo == 'hard':
        df_filtrado = aplicar_filtros_estrictos(df_original, condiciones_usuario)
        return df_filtrado if not df_filtrado.empty else "No se encontraron coincidencias exactas."

    elif modo == 'soft':
        vector_usuario = vectorizar_condiciones(condiciones_usuario, df_ml, columnas_modelo)
        distancias, indices = knn.kneighbors(vector_usuario, n_neighbors=k)
        return df_original.iloc[indices[0]]

    else:
        return "Modo no válido. Usa 'hard' o 'soft'."
```
