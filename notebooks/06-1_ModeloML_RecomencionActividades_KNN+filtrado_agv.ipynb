He introducido una normalización, para aceptar mayúsculas/ minúsculas, con y sin acentos.

```python
# Funciones para normalizar 
import unicodedata
def normalizar(texto):
    if not isinstance(texto, str):
        return texto
    texto = texto.lower().strip()
    texto = unicodedata.normalize('NFD', texto)
    texto = ''.join(c for c in texto if unicodedata.category(c) != 'Mn')
    return texto
```

Ahora he de filtrar mi dataframe original según las condiciones estipuladas por el usuario/ cliente.

```python
# Función de filtrado de los datos según condiciones usuario
import pandas as pd

def filtrado_segun_condiciones(df, condiciones):
    df_filtrado = df.copy()

    # --- Función de normalización ---
    def normalizar(texto):
        return str(texto).strip().lower()

    # --- Validación obligatoria: ciudad ---
    if 'ciudad' not in condiciones or normalizar(condiciones['ciudad']) != 'madrid':
        print("📍 Ciudad no proporcionada o inválida. Te mostramos las actividades en Madrid, ciudad en la que estamos trabajando actualmente.")
        condiciones['ciudad'] = 'Madrid'
    df_filtrado = df_filtrado[df_filtrado['ciudad'].str.lower() == 'madrid']
    
    # --- Normalización de tipos de columnas ---
    df_filtrado['edad_min'] = pd.to_numeric(df_filtrado['edad_min'], errors='coerce')
    df_filtrado['edad_max'] = pd.to_numeric(df_filtrado['edad_max'], errors='coerce')
    fecha_inicio_dt = pd.to_datetime(df_filtrado['fecha_inicio'], errors='coerce', dayfirst = True)
    fecha_fin_dt = pd.to_datetime(df_filtrado['fecha_fin'], errors='coerce', dayfirst = True)
    df_filtrado['precio_valor'] = pd.to_numeric(df_filtrado['precio_valor'], errors='coerce')

    # --- Filtro por rango de edad ---
    if 'rango_edad' in condiciones and condiciones['rango_edad']:
        edad_min, edad_max = condiciones['rango_edad']
        df_filtrado = df_filtrado[
            (df_filtrado['edad_max'] >= edad_min) & (df_filtrado['edad_min'] <= edad_max)
        ]

    # --- Filtro por fecha específica ---
    if 'fecha' in condiciones and condiciones['fecha']:
        fecha_usuario = pd.to_datetime(condiciones['fecha'], dayfirst=True, errors='coerce')
        if pd.notna(fecha_usuario):
            df_filtrado = df_filtrado[
                (fecha_inicio_dt <= fecha_usuario) & 
                (fecha_fin_dt >= fecha_usuario)
            ]

    # --- Filtro por día ---
    if 'día' in condiciones and condiciones['día']:
        dia = normalizar(condiciones['día'])
        if dia == "fin de semana":
            df_filtrado = df_filtrado[df_filtrado['día_días'].apply(normalizar).isin(["sábado", "domingo"])]
        elif dia == "día de semana":
            df_filtrado = df_filtrado[~df_filtrado['día_días'].apply(normalizar).isin(["sábado", "domingo"])]
        else:
            df_filtrado = df_filtrado[df_filtrado['día_días'].apply(normalizar) == dia]
    
    # --- Filtro por periodicidad ---
    if 'periodicidad' in condiciones and condiciones['periodicidad']:
        periodicidad = normalizar(condiciones['periodicidad'])
        df_filtrado = df_filtrado[
            df_filtrado['periodicidad'].apply(lambda x: normalizar(str(x))) == periodicidad
        ]
        
    # --- Filtro por horario ---
    if 'horario' in condiciones and condiciones['horario']:
        horario = normalizar(condiciones['horario'])

        # Elimina nulos antes de convertir
        df_filtrado = df_filtrado[df_filtrado['hora_inicio'].notnull()].copy()

        def hora_a_decimal(hora_str):
            try:
                h, m = map(int, hora_str.split(':'))
                return h + m / 60
            except:
                return None

        df_filtrado['hora_decimal'] = df_filtrado['hora_inicio'].apply(hora_a_decimal)

        if horario == "mañana":
            df_filtrado = df_filtrado[(df_filtrado['hora_decimal'] >= 7) & (df_filtrado['hora_decimal'] < 12)]
        elif horario == "mediodía":
            df_filtrado = df_filtrado[(df_filtrado['hora_decimal'] >= 12) & (df_filtrado['hora_decimal'] < 15)]
        elif horario == "tarde":
            df_filtrado = df_filtrado[(df_filtrado['hora_decimal'] >= 15) & (df_filtrado['hora_decimal'] < 19)]
        elif horario == "noche":
            df_filtrado = df_filtrado[(df_filtrado['hora_decimal'] >= 19) | (df_filtrado['hora_decimal'] < 7)]

        df_filtrado = df_filtrado.drop(columns=['hora_decimal'], errors='ignore')


    # --- Filtro por estación ---
    if 'estación' in condiciones and condiciones['estación']:
        estacion = normalizar(condiciones['estación'])
        mes_inicio = fecha_inicio_dt.dt.month
        if estacion == 'verano':
            df_filtrado = df_filtrado[mes_inicio.isin([6, 7, 8, 9])]
        elif estacion == 'invierno':
            df_filtrado = df_filtrado[mes_inicio.isin([12, 1, 2])]

    # --- Filtro por duración ---
    if 'duración' in condiciones and condiciones['duración']:
        duracion_dias = (fecha_fin_dt - fecha_inicio_dt).dt.days + 1
        duracion = normalizar(condiciones['duración'])
        if duracion == "corta":
            df_filtrado = df_filtrado[duracion_dias <= 1]
        elif duracion == "media":
            df_filtrado = df_filtrado[(duracion_dias > 1) & (duracion_dias <= 7)]
        elif duracion == "larga":
            df_filtrado = df_filtrado[duracion_dias > 7]

    # --- Filtro por distrito ---
    if 'distrito' in condiciones and condiciones['distrito']:
        distrito = normalizar(condiciones['distrito'])
        df_filtrado = df_filtrado[df_filtrado['distrito'].apply(normalizar) == distrito]

    # --- Filtro por precio ---
    if 'precio' in condiciones and condiciones['precio']:
        precio = normalizar(condiciones['precio'])
        if precio == "gratuito":
            df_filtrado = df_filtrado[df_filtrado['precio_valor'] == 0]
        elif precio == "bajo":
            df_filtrado = df_filtrado[(df_filtrado['precio_valor'] > 0) & (df_filtrado['precio_valor'] < 10)]
        elif precio == "medio":
            df_filtrado = df_filtrado[(df_filtrado['precio_valor'] >= 10) & (df_filtrado['precio_valor'] <= 30)]
        elif precio == "alto":
            df_filtrado = df_filtrado[df_filtrado['precio_valor'] > 30]

    # --- Filtro por categoría ---
    if 'categoría' in condiciones and condiciones['categoría']:
        categoria = normalizar(condiciones['categoría'])
        df_filtrado = df_filtrado[df_filtrado['categoría'].apply(normalizar) == categoria]

    # --- Filtro por subcategoría ---
    if 'subcategoría' in condiciones and condiciones['subcategoría']:
        subcategoria = normalizar(condiciones['subcategoría'])
        df_filtrado = df_filtrado[df_filtrado['subcategoría'].apply(normalizar) == subcategoria]

    print(df_filtrado.shape)
    return df_filtrado
```

### Vectorización condiciones del usuario

```python
# Funciones de vectorización condiciones usuario (ChatGPT)
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import datetime
import unidecode

def normalizar_texto(texto):
    return unidecode.unidecode(texto.strip().lower()) if isinstance(texto, str) else texto

def convertir_fecha_numerica(fecha_str, referencia=None):
    if isinstance(fecha_str, datetime.date):
        return fecha_str
    try:
        return datetime.datetime.strptime(fecha_str, "%d/%m/%Y").date()
    except (ValueError, TypeError):
        return None

def vectorizar_condiciones(condiciones_usuario, columnas_df_ml):
    vector_usuario = pd.DataFrame(np.ones((1, len(columnas_df_ml)), dtype=int), columns=columnas_df_ml)

    if 'rango_edad' in condiciones_usuario:
        edad_min, edad_max = condiciones_usuario['rango_edad']
    else:
        edad_min, edad_max = 0, 18

    vector_usuario['edad_min'] = edad_min
    vector_usuario['edad_max'] = edad_max

    for col in columnas_df_ml:
        if col.startswith('edad_') and '_' in col and 'min' not in col and 'max' not in col:
            try:
                r1, r2 = map(int, col.split('_')[1:])
                vector_usuario[col] = int(edad_min <= r1 and edad_max >= r2)
            except ValueError:
                pass

    horarios_dict = {
        "mañana": (7, 12), "mediodía": (12, 15),
        "tarde": (15, 19), "noche": (19, 7)
    }

    if 'horario' in condiciones_usuario:
        horario_usuario = normalizar_texto(condiciones_usuario['horario'])
        if horario_usuario in horarios_dict:
            h_ini, h_fin = horarios_dict[horario_usuario]
            vector_usuario['hora_inicio'] = h_ini
            vector_usuario['hora_fin'] = h_fin
    else:
        vector_usuario['hora_inicio'] = 7
        vector_usuario['hora_fin'] = 7

    if 'día' in condiciones_usuario:
        dia_usuario = normalizar_texto(condiciones_usuario['día'])
        dias_dict = {
            "lunes": "dia_lunes", "martes": "dia_martes", "miércoles": "dia_miercoles",
            "jueves": "dia_jueves", "viernes": "dia_viernes", "sábado": "dia_sabado", "domingo": "dia_domingo",
            "día de semana": ["dia_lunes", "dia_martes", "dia_miercoles", "dia_jueves", "dia_viernes"],
            "fin de semana": ["dia_sabado", "dia_domingo"]
        }

        for val in dias_dict.values():
            if isinstance(val, list):
                for col in val:
                    vector_usuario[col] = 0
            else:
                vector_usuario[val] = 0

        if dia_usuario in dias_dict:
            cols_a_activar = dias_dict[dia_usuario]
            if isinstance(cols_a_activar, list):
                for col in cols_a_activar:
                    vector_usuario[col] = 1
            else:
                vector_usuario[cols_a_activar] = 1

    # Procesar la fecha proporcionada por el usuario
    hoy = datetime.date.today()

    if 'fecha' in condiciones_usuario:
        fecha_str = condiciones_usuario['fecha']
        fecha_dt = convertir_fecha_numerica(fecha_str, referencia=hoy)
        vector_usuario['fecha'] = fecha_dt if fecha_dt else hoy
    else:
        vector_usuario['fecha'] = hoy

    # Convertir a datetime
    fecha_dt = pd.to_datetime(vector_usuario.at[0, 'fecha'])

    # Asignar componentes a columnas existentes (si existen)
    if 'fecha_inicio_year' in vector_usuario.columns:
        vector_usuario['fecha_inicio_year'] = fecha_dt.year
    if 'fecha_inicio_month' in vector_usuario.columns:
        vector_usuario['fecha_inicio_month'] = fecha_dt.month
    if 'fecha_inicio_day' in vector_usuario.columns:
        vector_usuario['fecha_inicio_day'] = fecha_dt.day
    if 'fecha_inicio_dayofweek' in vector_usuario.columns:
        vector_usuario['fecha_inicio_dayofweek'] = fecha_dt.weekday()
    if 'fecha_inicio_is_weekend' in vector_usuario.columns:
        vector_usuario['fecha_inicio_is_weekend'] = int(fecha_dt.weekday() >= 5)
    if 'fecha_inicio_is_summer' in vector_usuario.columns:
        vector_usuario['fecha_inicio_is_summer'] = int(6 <= fecha_dt.month <= 9)
    if 'fecha_inicio_is_winter' in vector_usuario.columns:
        vector_usuario['fecha_inicio_is_winter'] = int(fecha_dt.month == 12 or fecha_dt.month <= 2)

    if 'fecha_fin_year' in vector_usuario.columns:
        vector_usuario['fecha_fin_year'] = fecha_dt.year
    if 'fecha_fin_month' in vector_usuario.columns:
        vector_usuario['fecha_fin_month'] = fecha_dt.month
    if 'fecha_fin_day' in vector_usuario.columns:
        vector_usuario['fecha_fin_day'] = fecha_dt.day
    if 'fecha_fin_dayofweek' in vector_usuario.columns:
        vector_usuario['fecha_fin_dayofweek'] = fecha_dt.weekday()
    if 'fecha_fin_is_weekend' in vector_usuario.columns:
        vector_usuario['fecha_fin_is_weekend'] = int(fecha_dt.weekday() >= 5)
    if 'fecha_fin_is_summer' in vector_usuario.columns:
        vector_usuario['fecha_fin_is_summer'] = int(6 <= fecha_dt.month <= 9)
    if 'fecha_fin_is_winter' in vector_usuario.columns:
        vector_usuario['fecha_fin_is_winter'] = int(fecha_dt.month == 12 or fecha_dt.month <= 2)

    # Eliminar la columna 'fecha' temporal usada
    vector_usuario = vector_usuario.drop(columns=['fecha'])


    # Escalar valores numéricos
    scale_cols = ['edad_min', 'edad_max', 'hora_inicio', 'hora_fin', 'precio_valor']
    scale_cols = [col for col in scale_cols if col in vector_usuario.columns]
    vector_usuario[scale_cols] = vector_usuario[scale_cols].astype(float)
    if scale_cols:
        scaler = StandardScaler()
        vector_usuario[scale_cols] = scaler.fit_transform(vector_usuario[scale_cols])

    return vector_usuario
```

### Función **Recomendar actividades**

```python
# v.2 función recomendar con modo 'soft' ponderando condiciones clave
# condiciones clave son 'ciudad''distrito''rango_edad''fecha'
def recomendar_actividades(condiciones_usuario, modo='hard', k=5, 
                           df_original=None, df_ml=None, 
                           modelo_knn_filtrado=None, modelo_knn_global=None,
                           df_filtrado=None, df_filtrado_ml=None):
    
    validacion = validar_condiciones_usuario(condiciones_usuario)
    if validacion is not True:
        return {"status": "error", "errores": validacion}

    try:
        # Usamos un vector de ejemplo para el usuario (esto debería venir de tu vectorización real)
        # Asegúrate de usar vectorizar_condiciones cuando tengas la implementación real
        vector_usuario = df_ml.iloc[[25]]  # Esto debe ser reemplazado con la vectorización real
    except Exception as e:
        return {"status": "error", "mensaje": f"Error al vectorizar condiciones: {e}"}

    if modo == 'hard':
        # Asegúrate de usar los datos filtrados y el modelo filtrado
        if df_filtrado_ml is None or modelo_knn_filtrado is None or df_filtrado is None:
            return {"status": "error", "mensaje": "Faltan parámetros necesarios para modo 'hard'."}
        
        # Verificar si hay suficientes actividades para recomendar
        if len(df_filtrado_ml) < k:
            mensaje = f"Solo hay {len(df_filtrado_ml)} actividades disponibles (menos que k={k}). Se muestran todas."
            k = len(df_filtrado_ml)
        else:
            mensaje = f"Modo: {modo} - Recomendación generada correctamente con {k} vecinos más cercanos."
            
        # Convertimos el vector del usuario en un DataFrame con las columnas correctas
        vector_df = pd.DataFrame([vector_usuario.values[0]], columns=df_ml.columns)
        
        # Obtener las distancias y los índices con el modelo filtrado
        distancias, indices = modelo_knn_filtrado.kneighbors(vector_df, n_neighbors=k)
        
        # Obtener los resultados usando los índices y añadir las distancias
        resultados = df_filtrado.iloc[indices[0]].copy()
        resultados["distancia"] = distancias[0]
        
        # Imprimir un mensaje de estado
        print(mensaje)
        
        # Obtener los resultados usando los índices y añadir las distancias
        resultados = df_filtrado.iloc[indices[0]].copy()
        resultados["distancia"] = distancias[0]
        
        # Retornar directamente el DataFrame con los resultados
        return resultados

    elif modo == 'soft':
        if df_ml is None or modelo_knn_global is None or df_original is None:
            return {"status": "error", "mensaje": "Faltan parámetros necesarios para modo 'soft'."}
        
        df_candidato = df_ml.copy()
        df_candidato_original = df_original.copy()
        condiciones = condiciones_usuario.copy()
        
        # Ciudad
        if condiciones.get("ciudad") == "Madrid":
            df_candidato = df_candidato[df_original["ciudad"] == "Madrid"]
            df_candidato_original = df_candidato_original[df_original["ciudad"] == "Madrid"]

        # Distrito
        if "distrito" in condiciones:
            distrito = condiciones["distrito"]
            df_candidato = df_candidato[df_original["distrito"] == distrito]
            df_candidato_original = df_candidato_original[df_original["distrito"] == distrito]

        # Rango de edad
        if "rango_edad" in condiciones:
            edad_min, edad_max = condiciones["rango_edad"]
            df_candidato = df_candidato[
                (df_original["edad_min"] <= edad_max) & 
                (df_original["edad_max"] >= edad_min)
            ]
            df_candidato_original = df_candidato_original[
                (df_original["edad_min"] <= edad_max) & 
                (df_original["edad_max"] >= edad_min)
            ]

        # Fecha
        if "fecha" in condiciones:
            fecha = condiciones["fecha"]
            df_candidato = df_candidato[df_original["fecha_inicio"] <= fecha]
            df_candidato = df_candidato[df_original["fecha_fin"] >= fecha]
            df_candidato_original = df_candidato_original[
                (df_original["fecha_inicio"] <= fecha) &
                (df_original["fecha_fin"] >= fecha)
            ]

        # Si hay suficientes actividades candidatas, usamos ese subconjunto
        if len(df_candidato) >= k:
            vector_df = pd.DataFrame([vector_usuario.values[0]], columns=df_ml.columns)
            distancias, indices = modelo_knn_global.kneighbors(vector_df, n_neighbors=k)
            resultados = df_candidato_original.iloc[indices[0]].copy()
            resultados["distancia"] = distancias[0]
            print(f"Modo: {modo} - Recomendación generada con filtrado suave y {k} vecinos.")
            return resultados

        # Fallback global si no hay suficientes
        vector_df = pd.DataFrame([vector_usuario.values[0]], columns=df_ml.columns)
        distancias, indices = modelo_knn_global.kneighbors(vector_df, n_neighbors=k)
        resultados = df_original.iloc[indices[0]].copy()
        resultados["distancia"] = distancias[0]
        print(f"Modo: {modo} - Recomendación generada sin filtrado (fallback global) con {k} vecinos.")
        return resultados

        
    else:
        # Si el modo no es válido, devolver un mensaje de error
        print("Modo no válido. Usa 'hard' o 'soft'.")
        return None
```
